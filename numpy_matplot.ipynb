```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)  # 100 float values between -10 to +10
sigmoid = 1 / (1 + np.exp(-x))  # sigmoid formula  e^x/(e^x+1)

#basic of ploting graph a sigmoid graph
plt.axvline(0, color='red', linestyle='--', label='x = 0')
plt.axhline(0.5, color='red', linestyle='--', label='y = 0.5')
plt.plot(x, sigmoid, label="Sigmoid Function")
plt.xlabel("x")
plt.ylabel("Sigmoid(x)")
plt.title("Sigmoid Curve")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# random graph generating with each run also Procedural generation of graph points 
y = np.random.randint(1, 100, size=10)

# X-axis values (0 to 9)
x = np.arange(10)
#to find mid using the max fxn 
x_mid = x.max() / 2
y_mid = y.max() / 2


# this is the lines dividing the plane into 4 parts

plt.axhline( y_mid, color='pink', linestyle='--')
plt.axvline(x_mid, color='pink', linestyle='--')

#the orange dot just for reference to know where is mid 
plt.scatter(x_mid,y_mid , color = 'darkorange',marker='o' , zorder = 100)
plt.text(x_mid +0.2 , y_mid  +0.6, s=f'({x_mid} , {y_mid})', fontsize=13, color='black' ,zorder=100)


# responsible for blue line 
plt.plot(x, y, marker='o', color='blue', linestyle='-')
for i in x:
    plt.axvline(i, color='brown', linestyle='--')

plt.title("Procedural Line Graph")
plt.xlabel("the plane ")
plt.ylabel("displacement from X axis ")
plt.grid(True)
plt.show()


```

```python
import numpy as np
import matplotlib.pyplot as plt

# Random Y values
y = np.random.randint(1, 100, size=10)
x = np.arange(len(y))  # x = [0, 1, ..., 9]

# Plot the points
plt.plot(x, y, marker='o', color='blue', linestyle='-')

# Optional: limit x-axis to stop exactly at the last point
plt.xlim(x[0], x[-1])  # from first to last x

# Title and labels
plt.title("Line Touching Points Only")
plt.xlabel("Index")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```
